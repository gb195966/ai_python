```python
try:
  pegando_algo = int(input("Digite a sua idade : "))
except ValueError:
  print("Voce nao digitou um numero inteiro valido")
```

```python
try:
  pessoas = 6
  pessoas_2 = "Mario, Joao, Silvia , Maria"
  pessoas + pessoas_2
except ValueError as ve:
  print("Voce nao digitou um numero inteiro valido")
  print("Detalhes do erro:",ve)
except TypeError as te:
  print("Voce nao pode somar um numero com uma string")
  print("Detalhes do erro:",te)
```

```python
try:
  pegando_algo = int(input("Digite a sua idade : "))
  magica = 25/pegando_algo
  print(magica)
except ValueError as ve:
  print("Voce nao digitou um numero inteiro valido")
  print("Detalhes do erro:",ve)
except TypeError as te:
  print("Voce nao pode somar um numero com uma string")
  print("Detalhes do erro:",te)
except Exception as e:#pega todos os erros
  print("Detalhes do erro:",e)
finally:
  print("Finalizei o processo")
```

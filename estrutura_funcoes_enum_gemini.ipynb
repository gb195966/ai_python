# Estrutura de repeticao "for"

```python

dict_desafio = [
    {"nome":"Maria Silva", "media":8.9},
    {"nome":"Joao Silva", "media":7.5},
    {"nome":"Ana Oliveira", "media":4.2},
    {"nome":"Pedro Costa", "media":1.4}
]
```

```python
n = 0

while n < len(dict_desafio):
  print(dict_desafio)
  n +=1
```

2. Criar uma funcao que vai receber como argumento/parametro essa lista de emails, percorrer essa lista de emails, e, para cada email da lista, ela irá chamar a IA fo Gemini via API e pedir para ela resumir para voce todos os emails em apenas 1 linha, definindo o que essa pessoa queria no email

```python
email_bodies = [
    "Olá equipe, gostaria de agendar uma reunião para alinharmos as próximas entregas do projeto.",
    "Gostaria de solicitar minhas férias para o período entre 10 e 20 de julho.",
    "Identificamos uma instabilidade no sistema de pagamentos durante a madrugada.",
    "Você está convidado para participar do nosso evento anual de tecnologia.",
    "Realizamos uma atualização nos servidores para melhorar a performance da aplicação.",
    "Agradecemos sua participação no processo seletivo. Em breve retornaremos com novidades.",
    "Seu pagamento foi processado com sucesso e já consta em nosso sistema.",
    "Por motivos de segurança, recomendamos a alteração da sua senha imediatamente.",
    "Segue em anexo o relatório consolidado referente ao último trimestre.",
    "Todos os colaboradores devem concluir o treinamento até sexta-feira.",
    "Gostaríamos de discutir oportunidades de parceria entre nossas empresas.",
    "O sistema ficará indisponível neste domingo das 02h às 05h para manutenção.",
    "Seja bem-vindo à empresa! Estamos felizes em ter você conosco.",
    "Precisamos revisar algumas cláusulas do contrato vigente.",
    "Sua opinião é muito importante. Responda nossa pesquisa de satisfação.",
    "Temos o prazer de anunciar o lançamento do nosso novo produto.",
    "Detectamos um erro na integração com o serviço externo e já estamos atuando.",
    "O orçamento enviado foi aprovado pela diretoria financeira.",
    "Recebemos uma solicitação de recuperação de acesso à sua conta.",
    "Seu chamado foi resolvido e será encerrado automaticamente em 48 horas."
]

print(email_bodies)
```

```python
import os
from google.colab import userdata

from google import genai
client = genai.Client()


os.environ["GOOGLE_API_KEY"] = userdata.get('GEMINI_API_KEY')

def resumidor_de_emails(lista_de_emails):
  for numero,email in enumerate(lista_de_emails):
    resposta = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=f"""Vou te mandar o corpo de um e-mail. Quero que voce o resuma em apenas 1 linha,
        "passando o intuito daquele e-mail. Segue o e-mail: {email}"""
        )
    print(f"Email{numero}: {resposta.text}")
    print("-" * 50)

```

```python
resumidor_de_emails(email_bodies)
```

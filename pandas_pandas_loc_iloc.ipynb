```python
from groq import Groq
client = Groq()
os.environ["GROOQ_API_KEY"] = userdata.get('GROOQ_API_KEY')
```

```python
def resumidor_de_emails(lista_de_emails):

  lista_de_resumos = []
  for numero,email in enumerate(lista_de_emails):
    resposta = client.chat.completions.create(
    model="openai/gpt-oss-20b",
    messages=[
      {
        "role": "user",
        "content":f"""Vou te mandar o corpo de um e-mail. Quero que voce o resuma em apenas 1 linha,
        passando o intuito daquele e-mail. Segue o e-mail: {email}"""
      }
    ],
    temperature=1, #0 - menos criativo possivel; 2- o mais criativo possivel
    max_completion_tokens=7000,
    top_p=1,
    reasoning_effort="medium",
    stream=False, # Alterado para False
    stop=None
)
    # Acessa o conteúdo da resposta de forma não-streaming
    summary_content = resposta.choices[0].message.content
    print(f"Email{numero}: {summary_content}")
    lista_de_resumos.append(f"Email{numero}: {summary_content}")
    print("-" * 50)
  return lista_de_resumos
```

```python
#Etapa 1
lista_de_perguntas = [
    "De que é feito o Sol?",
    "De que é feito o planeta Saturno?",
    "Qual é a galaxia mais aintga ja encontrada?",
    "Qual é a maior estrela já encontrada?",
    "Qual é a estrela mais próxima do Sol?",
    "De que é feito o planeta Vênus?",
    "De que é feito o planeta Terra?",
    "De que é feito o planeta Júpiter?",
    "De que é feito o planeta Mercúrio?",
    "De que é feito o planeta Plutão?"
]

with open("perguntas.txt", "w", encoding="utf-8") as arquivo:
  for pergunta in lista_de_perguntas:
    arquivo.write(pergunta + "\n")
```

```python
#Etapa 2
lista_desafio = []
with open("perguntas.txt", "r", encoding="utf-8") as arquivo:
  for linha in arquivo:
    lista_desafio.append(linha.strip())
print(lista_desafio)
```

```python
#Etapa 3
import os
from google.colab import userdata

# Correctly set the environment variable GROQ_API_KEY before initializing the client
os.environ["GROQ_API_KEY"] = userdata.get('GROQ_API_KEY') # Assuming the secret in Colab is named 'GROQ_API_KEY'

from groq import Groq
client = Groq()

lista_de_dicionarios_de_resposta = []
for pergunta in lista_desafio:
    resposta = client.chat.completions.create(
    model="openai/gpt-oss-20b",
    messages=[
      {
        "role": "user",
        "content":f"""Gere uma resposta muito sucinta para a pergunta: {pergunta}"""
      }
    ],
    temperature=1,
    max_completion_tokens=7000,
    top_p=1,
    reasoning_effort="medium",
    stream=False,
    stop=None
  )
    # The line below was dead code and has been removed.
    lista_de_dicionarios_de_resposta.append({"pergunta":pergunta, "resposta":resposta.choices[0].message.content})
```

```python
lista_de_dicionarios_de_resposta
```

```python
#Etapa 4
with open("respostas1.csv", "w", encoding="utf-8") as arquivo:
  arquivo.write("pergunta,resposta\n")
  for pergunta_dict in lista_de_dicionarios_de_resposta:
    arquivo.write(f"{pergunta_dict['pergunta']},{pergunta_dict['resposta']}\n")
```

```python
import random
import pandas as pd

# Gerar 50 linhas de dados
num_rows = 50

# Listas para gerar dados aleatórios
nomes_produtos_base = ["Monitor", "Teclado", "Mouse", "Câmera", "Fone de Ouvido", "Tablet", "Smartwatch", "Carregador", "Caixa de Som", "Roteador", "Impressora", "Webcam", "Microfone", "Notebook", "Desktop"]
qualificadores_produtos = ["Gaming", "Bluetooth", "Sem Fio", "Ultra HD", "Profissional", "Compacto", "Ergonômico", "Básico", "Premium", "Portátil"]
categorias = ["Eletrônicos", "Informática", "Acessórios", "Áudio e Vídeo", "Periféricos", "Dispositivos Móveis"]

# Listas para armazenar os dados gerados
nome_do_produto = []
categoria_do_produto = []
preco_do_produto = []
quantos_itens_vendidos = []
avaliacao_do_produto = []

for _ in range(num_rows):
    # Nome do Produto
    nome_base = random.choice(nomes_produtos_base)
    qualificador = random.choice(qualificadores_produtos)
    nome_do_produto.append(f"{nome_base} {qualificador}")

    # Categoria do Produto
    categoria_do_produto.append(random.choice(categorias))

    # Preço do Produto (entre 10.00 e 1000.00)
    preco_do_produto.append(round(random.uniform(10.00, 1000.00), 2))

    # Quantos itens do produto foram vendidos (entre 10 e 500)
    quantos_itens_vendidos.append(random.randint(10, 500))

    # Avaliação do Produto (entre 1.0 e 5.0)
    avaliacao_do_produto.append(round(random.uniform(1.0, 5.0), 1))

# Criar o DataFrame
df_produtos = pd.DataFrame({
    'Nome do Produto': nome_do_produto,
    'Categoria do Produto': categoria_do_produto,
    'Preço do Produto': preco_do_produto,
    'Itens Vendidos': quantos_itens_vendidos,
    'Avaliação do Produto': avaliacao_do_produto
})

print(df_produtos.head())
```

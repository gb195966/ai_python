# Crie um codigo que percorra a lista de medias e some 1 a cada valor, exceto quando o valor passar de 10, que ai ele deve parar em 10.0

```python
n = 0

while n < 5:
  lista_de_medias[n] = lista_de_medias[n] + 1.0
  if lista_de_medias[n] > 10.0:
    lista_de_medias[n] = 10.0
  n = n + 1

print(lista_de_medias)
```

# Estrutura de repeticao "for"

```python

dict_desafio = [
    {"nome":"Maria Silva", "media":8.9},
    {"nome":"Joao Silva", "media":7.5},
    {"nome":"Ana Oliveira", "media":4.2},
    {"nome":"Pedro Costa", "media":1.4}
]
```

```python
n = 0

while n < len(dict_desafio):
  print(dict_desafio)
  n +=1
```

```python
for elemento in dict_desafio:
  print(elemento)
```

```python
for qualquer_nome in dict_desafio:
  print(qualquer_nome)
```

```python
for n in range(5):
  print(n)
```

```python
for n in range(2,7,3):
  print(n)
```

```python
pessoa = {
    "nome": "Fabricio",
    "idade": 19,
    "altura": 1.87
}
print(pessoa)
```

```python
for chave in pessoa.items():
  print(chave)
```

```python
for chave,valor in pessoa.items():
  print(f"Achave é {chave} e o valor é {valor} .")
```

```python
for n in range(11):
  if n%2 ==0:
    print(n)
```

```python
for n in range(11):
  if n%3 ==0:
    print(n)
```
